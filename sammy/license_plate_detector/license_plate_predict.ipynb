```python
# este notebook es para google colab
import zipfile
import os
from google.colab import files
from ultralytics import YOLO

# Define the directory and the zip file
folder_to_zip = "/content/runs"
zip_file_path = "/content/runs.zip"

# Create a zip file
with zipfile.ZipFile(zip_file_path, 'w') as zipf:
    for root, dirs, files_in_folder in os.walk(folder_to_zip):
        for file in files_in_folder:
            full_path = os.path.join(root, file)
            relative_path = os.path.relpath(full_path, folder_to_zip)
            zipf.write(full_path, relative_path)

# Download the zip file
files.download(zip_file_path)
```

```python
# Load your custom model
model2 = YOLO("/content/drive/MyDrive/la unitê/ITAM/2024/otoño/CDA I/yolo_weights/runs/detect/train3/weights/best.pt")  # Replace with your custom weights file
```

```python
import cv2
import matplotlib.pyplot as plt
# Load the image
image_path = "/content/drive/MyDrive/la unitê/ITAM/2024/otoño/CDA I/2398H9.jpg"
image = cv2.imread(image_path)

# Run the model prediction on the image
results = model2.predict(image)

# Show the prediction results
annotated_image = results[0].plot()

# Display the image with the detections
plt.imshow(annotated_image)
plt.axis('off')  # Hide axis
plt.show()
```
